```python
import pandas as pd
import numpy as np
import matplotlib.style as style
import matplotlib.pyplot as plt
```

```python
cod = pd.read_csv("./coldwellbankerMainData1.csv")
cod.head()
```

```python
cod.dtypes
```

```python
# renaming columns
cod.rename(columns = {"Price($)" : "Price_dollar", "Sq. Mt.":"Sq_Mt", "Price":"Price_Afl", "Sq. Ft.":"Sq_Ft", "Ref#":"Ref"}, inplace = True)
```

```python
cod.head()
```

```python
# correcting mispelt values in respective columns
# removing the characters from all values in a column
cod["Price_Afl"] = cod["Price_Afl"].str[4:]
```

```python
cod.head()
```

```python
cod["Sq_Mt"].tolist()
```

```python
cod["Sq_Mt"] = cod["Sq_Mt"].replace("30000 + 12090 m2", "42090")
```

```python
cod["Floors"] = cod["Floors"].replace("1 floor", "1")
```

```python
cod["Sq_Mt"].tolist()
```

```python
# correcting mispelt values
cod["Sq_Mt"] = cod["Sq_Mt"].map(lambda x:x.rstrip("m2M2 "))
```

```python
cod.head()
```

```python
cod["Floors"].tolist()
```

```python
cod["Sq_Mt"].tolist()
```

```python
# changing data types of respective columns
dt = {"Price_Afl":int, "Sq_Mt":float}
cod = cod.astype(dt)
```

```python
cod.dtypes
```

```python
# removing null values
cod = cod.dropna(subset=["Sq_Ft"])
cod.head()
```

```python
code = cod.groupby(["Property_name"])
a = code["Price_dollar"].sum().rename_axis("Property_name").reset_index(name = "prices")
a 
```

```python
a.sort_values(by= "prices",ascending= False, inplace=True)
a 
```

```python
b = a.nlargest(n = 10, columns= ["prices"])
b
```

```python
f = a.nsmallest(n = 10, columns= ["prices"])
f
```

```python
so = f.sort_values(by= "prices",ascending= False, inplace=True)
so
```

```python
so
```
